# Memory Usage Backend Comparisson

This experiment aims to evaluate the different measurements for the memory usage of the envelope seismic attribute operator.
On this notebook you will find:
- The problem statement
- The data collection for the experiment
- The evaluation of the experiment results.

## Problem statement

In order to evaluate the memory usage of the envelope seismic attribute operator, we need to compare the memory usage of the different backends available.
The backends available are the following:
- [resource](https://docs.python.org/3/library/resource.html)
- [psutil](https://psutil.readthedocs.io/en/latest/)
- [tracemalloc](https://docs.python.org/3/library/tracemalloc.html)
- [Direct kernel /proc file system](https://man7.org/linux/man-pages/man5/proc.5.html)

After executing the experiment, we expect the measurements to be consistent and the memory usage to be similar across the different backends.

## Evaluating Experiment Results

```python
from pathlib import Path; files = [str(p) for p in Path('your_directory').rglob('*.prof')]
from traceq import load_profile

profile_paths = [str(p) for p in Path(OUTPUT_DIR).rglob('*.prof')]
profile_names = [p.split('/')[-1].split('.')[0].split('-')[-1] for p in profile_paths]
profiles = [load_profile(p) for p in profile_paths]

zipped_profiles = list(zip(profile_names, profiles))

print(profile_paths)
```

Now, lets parse the memory and execution times

```python
profile_memory_usage_history = [
    (
        name,
        profile['metadata'][f'{name}_memory_usage_unit'],
        [entry[f'{name}_memory_usage'] for entry in profile['data']],
    ) for name, profile in zipped_profiles
]
profile_execution_time = [
    (
        name,
        profile['metadata']['unix_timestamp_unit'],
        profile['data'][-1]['unix_timestamp'] - profile['data'][0]['unix_timestamp'],
    ) for name, profile in zipped_profiles
]

print(profile_memory_usage_history[0])
print(profile_execution_time)
```

Now, lets create the plots

```python
import plotly.graph_objects as go

KB_TO_GB = 1048576
B_TO_GB = 1073741824

x_values_memory = []
y_values_memory_gb = []

fig1 = go.Figure()

for tool_name, unit, values in profile_memory_usage_history:
    if unit == 'kb':
        y_values_memory_gb = [value / KB_TO_GB for value in values]
    elif unit == 'b':
        y_values_memory_gb = [value / B_TO_GB for value in values]
    
    x_values_memory = list(range(len(values)))
    
    fig1.add_trace(go.Scatter(
        x=x_values_memory, 
        y=y_values_memory_gb, 
        mode='lines', 
        name=tool_name,
    ))

fig1.update_layout(
    xaxis_title="Time (log index)",
    yaxis_title="Memory Usage (GB)",
    xaxis=dict(showgrid=True, zeroline=True),
    yaxis=dict(showgrid=True, zeroline=True, exponentformat="E"),
    font=dict(family="Courier New, monospace", size=18, color="Black"),
    template="plotly_white",
    showlegend=True
)

fig1.write_image(f'{OUTPUT_DIR}/memory-usage-over-time.pdf', format="pdf", engine="kaleido")
fig1.show()
```

```python
import plotly.graph_objects as go

sorted_profile_execution_time = sorted(profile_execution_time, key=lambda x: x[2])
x_values_execution = [item[0] for item in sorted_profile_execution_time]
y_values_execution = [item[2] / 1000 for item in sorted_profile_execution_time]

fig2 = go.Figure()
fig2.add_trace(go.Bar(x=x_values_execution, y=y_values_execution))

fig2.update_layout(
    xaxis_title="Backend",
    yaxis_title="Execution Time (s)",
    font=dict(family="Courier New, monospace", size=18, color="Black"),
    template="plotly_white"
)

fig1.write_image(f'{OUTPUT_DIR}/execution-time-for-backend.pdf', format="pdf", engine="kaleido")
fig2.show()
```
